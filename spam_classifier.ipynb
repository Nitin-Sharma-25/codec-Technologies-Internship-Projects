```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. Dataset Initialization (Sample Data for Training)
data = {
    'text': [
        'Win a free lottery ticket now! Click here',
        'Hello, can we meet for coffee tomorrow?',
        'Get cheap loans with zero interest rate!',
        'Please review the attached project report.',
        'Urgent: Claim your cash prize of $1000'
    ],
    'label': ['spam', 'ham', 'spam', 'ham', 'spam']
}
df = pd.DataFrame(data)

# 2. Feature Extraction (Converting text into numerical vectors)
cv = CountVectorizer()
X = cv.fit_transform(df['text'])
y = df['label']

# 3. Model Training (Using Naive Bayes Algorithm)
model = MultinomialNB()
model.fit(X, y)

print("--- PROJECT 1: SPAM CLASSIFIER READY ---\n")

# 4. Interactive Testing Loop for Users
while True:
    user_input = input("Enter any email text to classify (or type 'exit' to quit): ")
    if user_input.lower() == 'exit':
        break

    # Transform user input and predict the classification label
    sample = cv.transform([user_input])
    prediction = model.predict(sample)[0]

    print(f"Result: This email is classified as **{prediction.upper()}**!\n")
```
